```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
import numpy as np
```

```python
input_cols = ['ANGER', 'CONTEMPT', 'DISGUST', 'FEAR', 'HAPPINESS', 'NEUTRAL', 'SADNESS', 'SURPRISE', 'ValenceMean-IAPS',
              'ArousalMean-IAPS', 'ValenceMean_IADS', 'ArousalMean_IADS']
output_cols = ['EXTRAVERSION',"OPENNESS","CONSCIENTIOUSNESS","AGREEABLENESS","NEUROTICISM" ]

dataset = pd.read_csv("preprocessed/ALL-DATA.csv")

data = dataset[input_cols]

data.describe()
```

```python
scaler_x = preprocessing.MinMaxScaler()
xscale = scaler_x.fit_transform(dataset)
```

```python
data.hist()
plt.savefig("histogram.png")
```

```python
all_columns = input_cols.copy()
all_columns.extend(output_cols)

dataset_2 = dataset[all_columns]
```

```python
correlations = dataset_2.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, annot=True,fmt='.2f')
fig.colorbar(cax)
ticks = np.arange(0,17,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(dataset_2.columns)
ax.set_yticklabels(dataset_2.columns)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = dataset_2.corr()
plt.figure(figsize=(10,8))
ax = sns.heatmap(correlation_matrix, vmax=1, square=True, annot=True,fmt='.2f', cmap ='GnBu', cbar_kws={"shrink": .5}, robust=True)
plt.title('Correlation matrix between the features', fontsize=20)
plt.show()
plt.savefig("bledny_histogram.png")
```
