```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from keras.layers import Dense, Dropout
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python

input_cols = [
#     'anger', 'contempt', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise', 
#     'picture_valence', 'picture_arousal', 
#     'sound_valence', 'sound_arousal', 
#     'ans_time', 
    'emotion_ans', 'valence_ans','arousal_ans',
    'picture_valence_diff', 'picture_arousal_diff',
    'sound_valence_diff', 'sound_arousal_diff',
#     'extraversion',"openness","conscientiousness","agreeableness","neurocism" 
]

dataset = pd.read_csv("preprocessed/ALL-DATA-2.csv")


print("Dataset isnull", dataset.isnull().any())

```

```python
x = dataset[input_cols].values
# scaler_x = preprocessing.MinMaxScaler()
# X_scale = scaler_x.fit_transform(x)

```

```python

kmeans = KMeans(n_clusters=4, random_state=0).fit(x)

y_kmeans = kmeans.predict(x)
```

```python
plt.scatter(x[:, 3], x[:, 4], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
plt.savefig("cluster-picture-val-arou.png")
```

```python
plt.scatter(x[:, 5], x[:, 6], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
plt.savefig("cluster-sound-val-arou.png")
```
